```python
import pandas as pd

flag = pd.read_csv("flag.csv")
flag.head()
```

```python
import altair as alt
bars = alt.Chart(flag).mark_bar().encode(
    x='country',
    y='landmass:N',
    color = 'zone:N'

)
bars
```

```python
brush = alt.selection_interval()
```

```python
bars = alt.Chart(flag).mark_bar().encode(
    x='country',
    y='landmass:N',
    color = 'landmass:N'

).transform_filter(
    brush
)
```

```python
points = alt.Chart(flag).mark_circle().encode(
    x='country',
    y='religion',
    color= 'religion:N',
    tooltip=['country', 'religion', 'landmass', 'language'],
   
    

).add_selection(
    brush
)

points
```

```python
points & bars
```

My Altair data dashboard uses the flag data set which provide information on different countries
The barchart looks at which continent(landmass) a country from. the continent are colour coded for quick data reading
The Scatter plot graph looks at the religion associated with each country aswell as providing information on each country when a specific country is selected.
